```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# graph for 2020 

df_20 = pd.read_csv('/Users/ironhack/Downloads/new_df2.csv')
df_20
```

```python
plt.figure(figsize=(10, 6))

# Define custom colors
custom_colors = {
    'DT Polls Sept20': '#f66868',  # Dark Red for DT polls
    'DT Results 2020': '#C8102E',   # Light Red for DT results
    'JB Polls Sept20': '#8282f3',   # Dark Blue for HC polls
    'JB Results 2020': '#0033A0',    # Light Blue for HC results
}

# Reshape the data to long format for easier plotting with seaborn
df_long = pd.melt(df_20, id_vars="state", value_vars=["DT Polls Sept20", "JB Polls Sept20", "DT Results 2020", "JB Results 2020"],
                  var_name="Metric", value_name="Percentage")

# Plot a bar chart
sns.barplot(x="state", y="Percentage", hue="Metric", data=df_long, palette=custom_colors)

# Rotate the x-axis labels for readability
plt.xticks(rotation=45, ha="right")

# Add title and labels
plt.title("2020 U.S. Presidential Election: Polls vs Results", fontsize=16)
plt.xlabel("State")
plt.ylabel("Percentage")
```
