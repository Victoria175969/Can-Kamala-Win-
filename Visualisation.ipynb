```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# graph for 2016 
df = pd.read_csv('/Users/ironhack/Downloads/new_df.csv')
df
```

```python
df_national = df[df['state'] == 'National']
df_others = df[df['state'] != 'National']
df_reordered = pd.concat([df_others, df_national], ignore_index=True)
df= df_reordered
df
```

```python
plt.figure(figsize=(10, 6))

# Define custom colors
custom_colors = {
    'DT Polls Sept16': '#f66868',  # Dark Red for DT polls
    'DT Results 2016': '#C8102E',   # Light Red for DT results
    'HC Polls Sept16': '#8282f3',   # Dark Blue for HC polls
    'HC Results 2016': '#0033A0',    # Light Blue for HC results
}

# Reshape the data to long format for easier plotting with seaborn
df_long = pd.melt(df, id_vars="state", value_vars=["DT Polls Sept16", "HC Polls Sept16", "DT Results 2016", "HC Results 2016"],
                  var_name="Metric", value_name="Percentage")

# Plot a bar chart
sns.barplot(x="state", y="Percentage", hue="Metric", data=df_long, palette=custom_colors)

# Rotate the x-axis labels for readability
plt.xticks(rotation=45, ha="right")

# Add title and labels
plt.title("2016 U.S. Presidential Election: Polls vs Results", fontsize=16)
plt.xlabel("State")
plt.ylabel("Percentage")
```
